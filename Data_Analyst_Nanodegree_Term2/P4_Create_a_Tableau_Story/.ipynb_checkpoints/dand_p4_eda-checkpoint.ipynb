```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import statsmodels.api as sm

%matplotlib inline
```

```python
df = pd.read_csv('2016.csv')
```

```python
df.info()
```

```python
df = df.set_index(['Country Name'])
```

```python
df = df.drop(['Country Code', 'Series Code'], axis=1)
```

```python
df = df.dropna()
```

```python
df = df.rename(index=str, columns={'2016 [YR2016]': '2016'})
```

```python
df.head()
```

```python
df['2016'] = df['2016'].replace('..', 0)
```

```python
df['2016'] = df['2016'].astype('float')
```

```python
df.info()
```

```python
df = df.pivot_table(index='Country Name', columns='Series Name', values='2016')
```

```python
df.head()
```

```python
df.to_csv('2016_cleaned.csv')
```
